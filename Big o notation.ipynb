```python
import matplotlib.pyplot as plt
import numpy as np

# Values
x = np.linspace(1,10,400)
x_factorial = np.arange(1,9) # to avoid overflow for factorial

# Functions
y1 = np.ones_like(x) # O(1)
y2 = np.log(x) # O(log n)
y3 = x # O(n)
y4 = x * np.log(x) # O(n log n)
y5 = x**2 # O(n^2)
y6 = 2**x # O(2^n)
y7 = [np.math.factorial(i) for i in x_factorial] # O(n!) - only calculated up to 8 to avoid overflow
y8 = x**x # O(n^n)

# Plotting
plt.figure(figsize=(10, 7))

plt.plot(x, y1, label='O(1)')
plt.plot(x, y2, label='O(log n)')
plt.plot(x, y3, label='O(n)')
plt.plot(x, y4, label='O(n log n)')
plt.plot(x, y5, label='O(n^2)')
plt.plot(x, y6, label='O(2^n)')
plt.plot(x_factorial, y7, label='O(n!)')
plt.plot(x, y8, label='O(n^n)')

plt.ylim(0,50) # limiting y values to 50 for better visualization
plt.xlim(0,10)

plt.xlabel('Elements')
plt.ylabel('Operations')
plt.title('Big O Notation')
plt.legend()
plt.grid(True)
plt.show()

```
