```python
#1.Use matplotlib to plot the following equation:

 y = x^2 -x +2
You'll need to write the code that generates the x and y points.

 Add an anotation for the point 0, 0, the origin.
```

```python
x = range(-100, 100)
y = [(x**2 - x + 2) for x in x]

# Plot x and y
plt.plot(x,y)

# add title
plt.title('$ y = x^2 - x +2$')

# add labels
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.annotate('Origin', xy = (0,0), xytext = (0, 3000), arrowprops ={})
%config InlineBackend.figure_format = 'png'
```

```python
#2.Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

y = âˆšx
y = x^3
y = 2^x
y = 1/(x+1)

#You can use functions from the math module to help implement some of the equations above.
```

```python
import matplotlib.pyplot as plt
import math

# We will create 4 subplots
fig, axs = plt.subplots(2, 2, figsize=(14,10))

# Values
x1 = range(0, 100)
x2 = range(0, 20)
x3 = range(0, 15)
x4 = [i*0.1 for i in range(0, 150)]  # We need floating point numbers here, so we increase the range and scale it down

# Calculate the values of y for each function
y1 = [math.sqrt(n) for n in x1]
y2 = [n**3 for n in x2]
y3 = [2**n for n in x3]
y4 = [1/(n+1) for n in x4]

# Plot each function
axs[0, 0].plot(x1, y1, color='pink')
axs[0, 0].set_title('$y=\sqrt{x}$')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('$\sqrt{x}$')
axs[0, 0].grid(True, ls = '--')

axs[0, 1].plot(x2, y2, color='purple')
axs[0, 1].set_title('$y=x^3$')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('$x^3$')
axs[0, 1].grid(True, ls = '--')

axs[1, 0].plot(x3, y3, color='cyan')
axs[1, 0].set_title('$y=2^x$')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('$2^x$')
axs[1, 0].grid(True, ls = '--')

axs[1, 1].plot(x4, y4, color='red')
axs[1, 1].set_title('$y=1/(x+1)$')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('$1/(x+1)$')
axs[1, 1].grid(True, ls = '--')

# To prevent the plots from overlapping
plt.tight_layout()
plt.show()
```

```python
# 3.Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points.Be sure to include a legend.

import matplotlib.pyplot as plt
import math

# Define the range of x for each function
x1 = [i*0.1 for i in range(0, 100)]
x2 = range(0, 20)
x3 = range(0, 10)
x4 = [i*0.1 for i in range(0, 100)]  # We need floating point numbers here, so we increase the range and scale it down

# Calculate the values of y for each function
y1 = [math.sqrt(n) for n in x1]
y2 = [n**3 for n in x2]
y3 = [2**n for n in x3]
y4 = [1/(n+1) for n in x4]

# Plot each function with a different color and label
plt.plot(x1, y1, label='$y=\sqrt{x}$', color='pink')
plt.plot(x2, y2, label='$y=x^3$', color='red')
plt.plot(x3, y3, label='$y=2^x$', color='cyan')
plt.plot(x4, y4, label='$y=1/(x+1)$', color='purple')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Combined plot')
plt.legend()
plt.grid(True, ls = '--')

plt.show()
```

```python
#4.Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.
import matplotlib.pyplot as plt
import numpy as np

# Define the range of x
x = np.linspace(0, 10, 400)

# Compute the values of y for each function
y1 = np.sqrt(x)
y2 = x**3
y3 = 2**x
y4 = 1/(x+1)

# Create a new figure
fig, ax = plt.subplots()

# Plot each function with a different color and label
ax.plot(x, y1, label='$y=\sqrt{x}$', color='pink')
ax.plot(x, y2, label='$y=x^3$', color='purple')
ax.plot(x, y3, label='$y=2^x$', color='cyan')
ax.plot(x, y4, label='$y=1/(x+1)$', color='red')

# Set the x and y axis labels
ax.set_xlabel('x')
ax.set_ylabel('y')

# Set the plot title and legend
ax.set_title('Combined plot of four functions')
ax.legend()

# Show the plot
plt.show()
```
